```python
import pandas as pd
import geopandas
import matplotlib
import matplotlib.pyplot as plt
```

```python
map = geopandas.read_file("./IlheusItab/IlheusItab.shp", encoding = 'utf-8')

map = map.drop(columns=["Id", "Lei_Orig", "Altera_1", "layer", "path"])
map = map.rename(columns={"CASOS": "Casos"})
```

```python
for i in range (len(map)):
    try:
        map.loc[i, "Casos"] = int(map.loc[i, "Casos"])
    except:
        map.loc[i, "Casos"] = 0
```

```python
plt.rcParams["figure.figsize"] = (10, 10)
```

```python
std_val = int(map["Casos"].std())
max_val = map["Casos"].max()
intervalo = [std_val, 2*std_val, 3*std_val, max_val]
intervalo
```

```python
map.plot(column="Casos", cmap='coolwarm', linewidth=1, legend=True)
```

```python
map.plot(column="Casos", scheme='user_defined', classification_kwds={'bins':intervalo}, cmap='coolwarm', linewidth=1, legend=True)
```
